```python
from vertexai.generative_models import GenerativeModel, GenerationConfig
from IPython.display import Markdown
import logging

# Instantiate flash_model with temperature=0
flash_model = GenerativeModel(
    model_name="gemini-2.0-flash",
    generation_config=GenerationConfig(
        temperature=0
    )
)

# Use the prompt_template and fill it with appropriate values
prompt_template = "# System_prompt\n{system_prompt} # Question\n{question} # Description {description}"

system_prompt = "You are an retail domestic merchandise expert"
question = "Provide a one sentence summary of the following text"

# Second product description (index 1)
description_text = description[1]

# Format the final prompt
final_prompt = prompt_template.format(
    system_prompt=system_prompt,
    question=question,
    description=description_text
)

# Generate content using Gemini flash model
llm_response = flash_model.generate_content(final_prompt)

# Display the Markdown-formatted output
display(Markdown(llm_response.text))

# Do not remove logging section
log_message = f"Markdown output: {llm_response.text}"
logging.info(log_message)
```

```python
import pandas as pd

# Prepare DataFrame for evaluation
dataset = pd.DataFrame(
    {
        "system_prompt": ["You are an retail domestic merchandise expert"] * len(description),
        "question": ["Provide a one sentence summary of the following text"] * len(description),
        "description": description,
    }
)
```

```python
from vertexai.evaluation import (
    MetricPromptTemplateExamples,
    EvalTask,
    PairwiseMetric,
    PairwiseMetricPromptTemplate,
    PointwiseMetric,
    PointwiseMetricPromptTemplate,
)
# # ✅ Fix: Pass required keyword argument correctly
# POINTWISE_METRIC = PointwiseMetric(
#     metric_prompt_template=MetricPromptTemplateExamples.Pointwise.SUMMARIZATION_QUALITY
# )

# Create the evaluation task
pointwise_eval_task = EvalTask(
    dataset=dataset,
    metrics=[MetricPromptTemplateExamples.Pointwise.GROUNDEDNESS],
    experiment="product-summarization-quality"
)

# Evaluate the model
pointwise_result = pointwise_eval_task.evaluate(
    model=flash_model,
    prompt_template="# System_prompt\n{system_prompt} # Question\n{question} # Description {description}",
)
```

```python
from vertexai.generative_models import GenerativeModel
from vertexai.evaluation import (
    MetricPromptTemplateExamples,
    PairwiseMetric,
    EvalTask
)

# ✅ Step 1: Instantiate the models
baseline_model = GenerativeModel(model_name="gemini-2.0-flash-lite")
candidate_model = GenerativeModel(model_name="gemini-2.0-flash")  # or use your existing `flash_model`

# ✅ Step 2: Properly instantiate the PairwiseMetric
pairwise_metric = PairwiseMetric(
    metric="summarization_quality",
    metric_prompt_template=MetricPromptTemplateExamples.Pairwise.SUMMARIZATION_QUALITY,
    baseline_model=baseline_model
)

# ✅ Step 3: Create the EvalTask (notice it's passed an **instance**, not the class)
pairwise_eval_task = EvalTask(
    dataset=dataset,
    metrics=[pairwise_metric],
    experiment="pairwise-product-summarization-quality"
)

# ✅ Step 4: Evaluate the candidate model vs. the baseline
pairwise_result = pairwise_eval_task.evaluate(
    model=candidate_model,  # or flash_model
    prompt_template="# System_prompt\n{system_prompt} # Question\n{question} # Description {description}"
)

```
